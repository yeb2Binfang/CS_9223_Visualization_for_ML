## 1号词云
4行代码简单上手

```
import wordcloud
import matplotlib.pyplot as plt
% matplotlib inline
```

```
w = wordcloud.WordCloud()
word = {'a':5,'b':7,'c':3}
w.generate_from_frequencies(word)
```

```
w.to_file('01-output.png')
```

```
plt.imshow(w)
plt.axis("off")
```

## 4号词云
中文分词 jieba

```
import jieba
```

中文分词jieba的参数

cut_all:表示把每个可能形成的词汇都提取出来

如果是lcut_for_search,这个就是将cut_all的分词从短到长排列好

colab不知道为什么不输出有中文的图片，所以，我们主要还是应用英文

```
textlist1 = jieba.lcut('动力学和电磁学')
print(textlist1)
```

```
textlist2 = jieba.lcut('动力学和电磁学',cut_all=True)
print(textlist2)
```

```
textlist3 = jieba.lcut_for_search('动力学和电磁学')
print(textlist3)
```

```
w = wordcloud.WordCloud(width=1000,height=700,background_color='white')
txt = '纽约大学（New York University），简称NYU，建立于1831年，由时任美国财政部长艾伯特·加勒廷创建，是一所位于美国纽约市的世界顶尖私立研究型大学，全美办学规模最大的私立名校之一，同时在伦敦、马德里、悉尼、柏林、巴黎等地共设立11个全球学术中心。'
##这两行非常重要，jieba 进行cut之后，是变成数组的，我们要把它变成string，所以需要使用join方法
txtlist = jieba.lcut(txt)
string = " ".join(txtlist)
```

```
w.generate(string)
```

```
plt.imshow(w)
plt.axis("off")
```

## 7号词云
不仅要按照图片模板走，而且词云的颜色要跟图片一样

```
from wordcloud import WordCloud, ImageColorGenerator
```

```
mk = imageio.imread("https://raw.githubusercontent.com/yeb2Binfang/CS_9223_Visualization_for_ML/main/data_visualizaiton_basis/Python_Basis/word_cloud/alice_color.png")
```

```
w = wordcloud.WordCloud(background_color='white'
                        ,mask=mk
                        ,scale=15)
```

```
f = open('word_cloud_3.txt',encoding='utf-8')
txt = f.read()
##这个就是普通的图，也就是各种颜色的
w_ori = w.generate(txt)
```

```
##这个就是把词云按照图片的颜色走
image_colors = ImageColorGenerator(mk)
w_color = w_ori.recolor(color_func=image_colors)
```

```
##

plt.subplot(1,2,1)
plt.imshow(w)
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(w_color,interpolation='bilinear')
plt.axis("off")
```
